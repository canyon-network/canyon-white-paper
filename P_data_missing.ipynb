```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

# x = minimal_storage_ratio
#
# y = number_of_validators
#
# P = probability_of_data_missing
#
# P = (1-x)^y < 0.0000000001

# Assuming P = 0.0000000001
alpha = math.pow(10, -10)


def calculate_y(x):
    y = 1
    while math.pow(1 - x, y) >= alpha:
        y += 1
    return y

x_list = np.linspace(0.1, 1.0, num = 100)

y_list = []

for x in np.linspace(0.1, 1.0, num = 100):
    y_list.append(calculate_y(x))

plt.figure('Scatter')

ax = plt.gca()

ax.set_xlabel('x = minimal storage ratio')
ax.set_ylabel('y = number of validators')
ax.set_title('The probability of data missing = 0.0000000001')

x_major_locator = plt.MultipleLocator(0.1)
y_major_locator = plt.MultipleLocator(20)

ax.xaxis.set_major_locator(x_major_locator)
ax.yaxis.set_major_locator(y_major_locator)

ax.scatter(x_list, y_list, c='r', s=20, alpha=0.5)

plt.grid()


plt.savefig("output.eps", format="eps")

plt.show()
```
